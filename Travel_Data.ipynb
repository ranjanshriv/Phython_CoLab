#Read Travel File#

```python
import pandas as pd
pd.options.display.float_format = '{:,.0f}'.format
#reading two sheets
df = pd.read_excel("/content/travel_data.xls",[0,1],3,usecols=['Country Code', '2016','2017'])
print(df[1])
#checking NaN value in column 2016 of 2nd sheet
df[1][~df[1]["2016"].isnull()]

#checking NaN value in all columns  of 2nd sheet  "~" using to reverse value result. e.g Isnull giving result True for NaN. ~using this it will False
df[1][~df[1].isnull().any(axis=1)]

#show all columns in dataframe
df[1].columns

#check datatype of all columns
df[1].dtypes

#select all columns with specific type
df[1].select_dtypes(include="float64")

# change index of Dataframe
df[1].index = df[1]["Country Code"]
df[1]
```
